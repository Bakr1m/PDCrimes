```python
cc = crime.groupby('DAY_OF_WEEK').count()['INCIDENT_NUMBER'].sort_values(ascending = False)
cc
```

```python
cc.plot(kind = 'bar', title = 'number of crimes by day')
```

```python
#on which days and hours crimes are committed
hour_and_day = crime.groupby(['HOUR', 'DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack()
hour_and_day.columns = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday',	'Sunday']
hour_and_day
```

```python
sns.heatmap(hour_and_day)
```

```python
#committed crimes by District on Yearly basis
district_and_year = crime.groupby(['YEAR','DISTRICT']).count()['INCIDENT_NUMBER'].unstack()
district_and_year
```

```python
district_and_year.plot(kind = 'bar', title = "committed crimes by District on Yearly basis")
```

```python
sns.heatmap(district_and_year)
```
